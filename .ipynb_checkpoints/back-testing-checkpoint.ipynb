### Load Libraries

```python
import pandas as pd
from utils.util import get_data, plot_data
```

### Initialization

```python
date_start = '2012-06-20'
date_end = '2012-09-12'
stock = 'IBM'
file_name = 'out\\backtest.csv'
n_stocks = 5

dates = pd.date_range(date_start, date_end)
```

### Load Data

```python
orders = pd.read_csv(file_name, index_col=0, parse_dates=True)
orders
```

```python
prices_all = get_data([stock], dates)
df = prices_all[[stock]]
df = df.join(orders[['Order', 'Shares']])
df['port_val'] = 0.0
df.head()
```

```python
trading = False

for i, row in df.iterrows():

    if row['Order'] == 'BUY':
        trading = True
    
    if trading == True:
        df.loc[i, 'port_val'] = n_stocks * row[stock]
    
    if row['Order'] == 'SELL':
        trading = False
        
df['diff'] = df['port_val'].diff()
```

```python
ix_buy = df[df['Order'] == 'BUY'].index
df.loc[ix_buy, 'diff'] = 0.0
```

```python
ix_sell = df[df['Order'] == 'SELL'].index
df_temp = df['diff'].shift(-1)
df_temp.loc[ix_sell] = 0.0
df['diff'] = df_temp.shift(+1) 
```

```python
df['cum_val'] = df['diff'].cumsum()
df
```

```python
plot_data(df['diff'])
```
